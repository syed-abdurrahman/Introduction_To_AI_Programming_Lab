## **Aim of the Lab**

The aim of this experiment is to practice creating **Python lambda functions** for quick electrical calculations. ## **Lab Tasks**

You’re writing small analysis tools for an electrical lab. The goal is to create quick one-line functions to calculate adjusted power readings, filter noisy sensor data, sort components by rating, and combine multiple electrical values using **Python lambda functions**.


### **Task 1 — Basic Lambda for a Simple Electrical Formula**

Write a lambda function that takes a voltage value `V` and returns:

\[
V^2 + 2
\]

This represents a toy “adjusted voltage metric.”  
Test it with `V = 5`.


---

```python
# Task 1: Lambda for adjusted voltage
adjusted_voltage = lambda V: V**2 + 2

# Test
V = 5
print("Adjusted Voltage:", adjusted_voltage(V))
```

### **Task 2 — Lambda with Two Electrical Parameters**

Write a lambda that takes two resistances `R1` and `R2` and returns their **equivalent parallel resistance**:

Req = (R1 * R2) / (R1 + R2)

Test it with:

- `R1 = 100 Ω`  
- `R2 = 200 Ω`


```python
# Task 2: Lambda for equivalent parallel resistance
parallel_resistance = lambda R1, R2: (R1 * R2) / (R1 + R2)

# Test
R1 = 100
R2 = 200
print("Equivalent Parallel Resistance:", parallel_resistance(R1, R2), "Ω")
```
