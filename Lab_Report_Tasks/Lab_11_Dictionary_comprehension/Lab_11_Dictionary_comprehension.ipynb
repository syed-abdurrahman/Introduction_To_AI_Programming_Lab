## **Aim of the Lab**

The aim of this laboratory experiment is to practice using **dictionary comprehensions** in Python for creating, filtering, transforming, and manipulating dictionaries concisely. The lab demonstrates efficient ways to map, filter, and update dictionary data for real-world datasets.

## **Lab Tasks:** 
`Dictionary Comprehensions`

You are helping a small learning center clean and organize short classroom datasets.  
The tasks focus on **dictionary comprehensions** for short and readable transformations.

**Task:**  
- Create a list of current values.  
- Loop through the list and calculate the square of each value.  
- Print the squared values.

---

### **Task 1 — Create Dictionary from a List**

Given a list:  

```python
names = ["Ali", "Sara", "Bilal"]

````python
# Dictionary mapping names to their length
names = ["Ali", "Sara", "Bilal"]
name_lengths = {name: len(name) for name in names}
print("Name lengths:", name_lengths)
````

-------

### **Task 2 — Filter a Dictionary Using Comprehension**

Given a dictionary of scores:

```python
scores = {"Ali": 55, "Sara": 82, "Bilal": 40, "John": 90}


````python
# Filter students with scores >= 60
scores = {"Ali": 55, "Sara": 82, "Bilal": 40, "John": 90}
passed_students = {k: v for k, v in scores.items() if v >= 60}
print("Students who scored >= 60:", passed_students)
````

---------

### **Task 3 — Swap Keys and Values**

Given a dictionary:

```python
pairs = {"a": 1, "b": 2, "c": 3}

````python
# Swap keys and values
pairs = {"a": 1, "b": 2, "c": 3}
swapped_pairs = {v: k for k, v in pairs.items()}
print("Swapped dictionary:", swapped_pairs)
````

-----------

### **Task 4 — Transform Dictionary Values**

Given a dictionary of prices:

```python
prices = {"pen": 10, "book": 50, "bag": 100}


````python
# Apply 10% discount
prices = {"pen": 10, "book": 50, "bag": 100}
discounted_prices = {k: v * 0.9 for k, v in prices.items()}
print("Discounted prices:", discounted_prices)
````
