## **Aim of the Lab**

The aim of this experiment is to understand and practice using **Python dictionaries** to store, access, and manipulate structured data. The experiment focuses on creating a dictionary for a solar home system, retrieving specific values, counting fields, removing items, and accessing keys, values, and key-value pairs using the `.keys()`, `.values()`, and `.items()` methods.


# Lab Tasks:

You are working in a university electronics lab. You are asked to monitor a small solar home system installed on the lab roof. The system consists of:

- **a)** A solar panel  
- **b)** A battery  
- **c)** A load controller  
- **d)** A temperature sensor  

Each minute, the system sends a small dictionary of readings.

---



**Q1:** Create a dictionary `reading1` with 5 fields.


```python
reading1 = {
    "solar_power": 320,
    "battery_voltage": 12.5,
    "load_status": "ON",
    "panel_current": 8.5,
    "temperature": 36
}
print(reading1)
```

**Q2:** From `reading1`, print:

i. The solar power  
ii. The battery voltage  
iii. The load status


```python
print("Solar Power:", reading1["solar_power"])
print("Battery Voltage:", reading1["battery_voltage"])
print("Load Status:", reading1["load_status"])
```

**Q3:** Print how many fields the dictionary contains after modifications


```python
print("Number of fields:", len(reading1))
```

**Q4:** Remove the "panel_current" field


```python
reading1.pop("panel_current")
print(reading1)
```

**Q5:** Get Keys, Values, and Items


```python
print("Keys:", reading1.keys())
print("Values:", reading1.values())
print("Items:", reading1.items())
```
