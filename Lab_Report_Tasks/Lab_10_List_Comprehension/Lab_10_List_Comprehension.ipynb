### **Task 1 — Basic List Comprehension**

Create a new list of numbers from 0 to 9 using a list comprehension, without using a loop.

**Syntax:** `[expression for item in iterable]`


````python
# Basic list comprehension
numbers = [i for i in range(10)]
print("Numbers 0-9:", numbers)
````

-----------

### **Task 2 — Filtering Items with `if`**

Given a list:  

```python
nums = [1, 2, 3, 4, 5, 6]

Use a list comprehension with an if clause to keep only the even numbers.
 [expr for item in iterable if condition]


````python
# Keep only even numbers
nums = [1, 2, 3, 4, 5, 6]
even_nums = [n for n in nums if n % 2 == 0]
print("Even numbers:", even_nums)
````

------------

### **Task 3 — Conditional Expression Inside Comprehension**

Given a list:  

```python
nums = [1, 2, 3, 4]

Create a new list where even numbers become "even" and odd numbers become "odd".
expr_if_true if condition else expr_if_false inside comprehension


````python
# Conditional expression inside comprehension
nums = [1, 2, 3, 4]
labels = ["even" if n % 2 == 0 else "odd" for n in nums]
print("Even/Odd labels:", labels)
````

------

### **Task 4 — Simple Nested List Comprehension**

Given a list of lists:  

```python
data = [[1, 2], [3, 4]]

Flatten it into a single list using a nested list comprehension

````python
# Nested list comprehension to flatten list
data = [[1, 2], [3, 4]]
flat_list = [item for sublist in data for item in sublist]
print("Flattened list:", flat_list)
````
