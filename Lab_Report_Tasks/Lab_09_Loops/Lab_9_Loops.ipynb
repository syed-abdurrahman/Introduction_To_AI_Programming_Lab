## **Aim of the Lab**

The aim of this experiment is to calculate the **square of current readings** to determine power loss in an electrical circuit. This task reinforces the use of **loops** and **list operations** in Python for processing numerical data.

## **Lab Report Task:**
`Calculating Power Loss from Current Readings`

Engineers want to calculate the **square of each current reading** for power loss calculation using the formula: [I**2]

**Task:**  
- Create a list of current values.  
- Loop through the list and calculate the square of each value.  
- Print the squared values.

---

```python
# List of current readings (in amperes)
current_readings = [2, 5, 7, 3, 4]
```

```python
# Calculate the square of each current reading
squared_currents = []
for I in current_readings:
    squared_currents.append(I**2)
```

```python
# Print the squared values
print("Current readings:", current_readings)
print("Squared readings (I^2):", squared_currents)
```

## **Conclusion of the Lab**

In this experiment, a list of current readings was created and each value was squared using a loop. The squared values represent the contribution of each current to power loss. This lab reinforced the use of loops and arithmetic operations on lists in Python for electrical engineering calculations.
