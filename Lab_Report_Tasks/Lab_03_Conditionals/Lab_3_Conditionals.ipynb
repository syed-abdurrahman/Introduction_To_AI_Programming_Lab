## **Aim of the Lab**

The aim of this laboratory experiment is to check the value of a resistor before using it in a circuit in order to prevent a short circuit. The resistance value is stored in a variable and an **if–else conditional statement** is used to determine whether the resistance is zero. If the resistance is zero, the program identifies it as a **short circuit**; otherwise, it confirms that the circuit is **safe**.

## **Task:**
A resistor with zero value creates a short circuit, which can damage the circuit. Students need to test the resistance before use.
Check if the resistance value is zero. If yes, print "Short Circuit", else print "Safe".
Store resistance in a variable, use if-else.


```python
resistance = 0

if resistance == 0:
    print("Short Circuit")
else:
    print("Safe")
```

## **Conclusion of the Lab**

In this laboratory experiment, the resistance value was successfully tested using an **if–else condition**. The program correctly identified a resistance value of zero as a **short circuit**, while non-zero resistance values were marked as **safe**. This lab demonstrated the importance of conditional statements in preventing circuit damage and reinforced the practical use of basic decision-making logic in programming and electrical safety analysis.
