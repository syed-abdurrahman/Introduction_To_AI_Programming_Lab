## Practice Task 1: Load Demand Data Recording

**Question:**  
Create a CSV file to store hourly load demand (kW) with timestamps.

**Answer:**

```python
import csv

load_data = [
    ["Time", "Load (kW)"],
    ["08:00", 50],
    ["09:00", 60],
    ["10:00", 55]
]
```

```python
with open("Lab_13_load_demand.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(load_data)

print("Lab_13_Load demand data recorded.")
```

## Practice Task 2: Solar Energy Generation Logging

**Question:**  
Create a file to store daily solar energy generation data and append new values.

**Answer:**


```python
import csv

solar_data = [
    ["Date", "Energy Generated (kWh)"],
    ["2025-09-01", 35],
    ["2025-09-02", 38]
]
```

```python
with open("Lab_13_solar_generation.csv", "w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(solar_data)
```

```python
# Append new record
with open("Lab_13_solar_generation.csv", "a", newline="") as file:
    writer = csv.writer(file)
    writer.writerow(["2025-09-03", 40])

print("Solar generation data stored and updated.")
```

## Task: Organize and Process CSV Data

**Question:**  
Read `Lab_13_load_demand.csv`, count total number of recorded days, and display each record with a serial number.  
Do not count the header.

**Answer:**


```python
import csv

with open("Lab_13_load_demand.csv", "r") as file:
    reader = csv.reader(file)
    header = next(reader)
    
    count = 0
    for i, row in enumerate(reader, start=1):
        print(i, row)
        count += 1

print("Total recorded days:", count)
```

## Task: Write Organized Data to a New CSV File

**Question:**  
Create a new file named `organized_energy_data.csv` and write the processed data into it.  
The original file must remain unchanged.

**Answer:**


```python
import csv

with open("Lab_13_load_demand.csv", "r") as infile, open("Lab_13_organized_energy_data.csv", "w", newline="") as outfile:
    reader = csv.reader(infile)
    writer = csv.writer(outfile)
    
    header = next(reader)
    writer.writerow(["S.No"] + header)
    
    for i, row in enumerate(reader, start=1):
        writer.writerow([i] + row)

print("Organized data written to organized_energy_data.csv")
```
