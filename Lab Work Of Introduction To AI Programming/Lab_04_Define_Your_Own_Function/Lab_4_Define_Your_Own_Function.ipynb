## 3. Solar Battery Charging:
A solar inverter monitors the batteryâ€™s state of charge (SOC).

Write a function charge_from_solar(SOC) that:
	`Prints "Charging" if SOC < 20%.
    
	`Prints "Battery full, no action" if SOC > 90%.
    
	`Prints "Battery charging, SOC = ---%" otherwise.
Add docstrings and function annotations.
Print the functionâ€™s docstring to check.

```python
def charge_from_solar(SOC: float) -> None:
    """
    Determine battery charging status.
    """
    if SOC < 20:
        print("Charging")
    elif SOC > 90:
        print("Battery full, no action")
    else:
        print(f"Battery charging, SOC = {SOC}%")

charge_from_solar(15)
charge_from_solar(60)
charge_from_solar(95)

print(charge_from_solar.__doc__)
```

```python
def charge_from_solar(SOC:float) -> float:
    """
    Check the battery State of Charge (SOC) and gives instructions.

    Parameters:
        SOC (float): State of Charge in percentage (0â€“100).

    Prints:
        - "Charging" if SOC < 20%
        - "Battery full, no action" if SOC > 90%
        - "Battery charging, SOC = ---%" otherwise
    """
    if SOC < 20:
        print("Charging")
    elif SOC > 90:
        print("Battery full, no action")
    else:
        print(f"Battery charging, SOC = {SOC}%")
print("\nFunction Docstring")
print(charge_from_solar.__doc__)
```

## 4. Resistance, Voltage, and Current Lists:
You are building a program to analyze five resistor values in a circuit.

Function 1: resistance(): return a list of 5 resistances.

Function 2: voltage(): return voltages across each resistor.

Function 3: current(): take both lists (Resistance & Voltage) and calculate currents.

Display all lists nicely in the main program.

Try two versions:

	`Return as a tuple.`
    
    `Return as a dictionary with keys like "current_1".`

```python
def resistance() -> list:
    return [10, 20, 30, 40, 50]

def voltage_list() -> list:
    return [5, 10, 15, 20, 25]

def current(R: list, V: list) -> tuple:
    return tuple(v/r for v, r in zip(V, R))

R = resistance()
V = voltage_list()
I = current(R, V)

print("Resistance:", R)
print("Voltage:", V)
print("Current (tuple):", I)
```

```python
def current_dict(R: list, V: list) -> dict:
    return {f"current_{i+1}": V[i]/R[i] for i in range(len(R))}

print("Current (dictionary):", current_dict(R, V))
```

## 6. Parallel Resistance:
Now, calculate parallel resistance.

Write a function parallel() that takes a list or tuple of resistances.

Return the equivalent resistance.

Add docstrings and annotations.

Print the functionâ€™s __annotations__.

```python
def parallel(resistances):
        # 1: empty input
    if not resistances:
        return 0.0

    # 2: short-circuit check
    if any(r == 0 for r in resistances):
        return 0.0

    # 3: compute reciprocal sum
    reciprocal_sum = 0.0
    for r in resistances:
        reciprocal_sum += 1.0 / r

    # 4: final result (handle zero reciprocal_sum)
    return 1.0 / reciprocal_sum if reciprocal_sum != 0.0 else float('inf')
def parallel(resistances: list[float] | tuple[float, ...]) -> float:
    """
    Calculate equivalent resistance for resistors in parallel.

    Parameters:
        resistances (list[float] | tuple[float, ...]): resistor values in ohms

    Returns:
        float: equivalent resistance in ohms.
               - returns 0.0 if list is empty or any resistor == 0 (short).
               - returns float('inf') if reciprocal sum is exactly zero.
    """
print(parallel.__annotations__)
```

```python
print("parallel([10, 20, 30]) ->", parallel([10, 20, 30]), "ohms")
print("parallel([]) ->", parallel([]), "ohms")
print("parallel([0, 10]) ->", parallel([0, 10]), "ohms")
```

```python
import torch

# Function to compute derivative using autograd
def dz_dx_autograd(x_value):
    x = torch.tensor(x_value, requires_grad=True)
    y = x**2
    z = torch.sin(y)
    z.backward()         # compute gradient
    return x.grad.item() # return derivative
x_values = [0.5, 1.0, 2.0]

print("x\tManual Derivative\tAutograd Derivative")
for val in x_values:
    manual = dz_dx(val)
    auto = dz_dx_autograd(val)
    print(f"{val}\t{manual:.6f}\t\t{auto:.6f}")
```

## Task 1:
Function with All Types of Arguments:

```python
def f(a, b, *args, **kwargs):
    print(f'a = {a}')
    print(f'b = {b}')
    print(f'args = {args}')
    print(f'kwargs = {kwargs}')
f(1, 2, 'foo', 'bar', 'baz', 'qux', x=100, y=200, z=300)
```

In this example, a and b are normal positional arguments passed to the function. Any extra positional values are automatically collected in args, which is a tuple. Similarly, all named (keyword) arguments are stored in kwargs, which is a dictionary. This allows the function to handle a variable number of positional and keyword arguments in a flexible way.

## Task 2:
Function with Only *args:

```python
def f(*args):
    for i in args:
        print(i)
a = [1, 2, 3]
t = (4, 5, 6)
s = {7, 8, 9}
f(*a, *t, *s)
```

We observe that you can unpack lists, tuples, and sets into a function. When doing so, all the unpacked values are collected by *args into a single tuple. This allows the function to handle multiple values in a flexible and organized manner.

## Task 3:
Function with Only **kwargs:


```python
def f(**kwargs):
    for k, v in kwargs.items():
        print(k, '->', v)
d1 = {'a': 1, 'b': 2}
d2 = {'x': 3, 'y': 4}
f(**d1, **d2)
```

We observe that **kwargs collects all named (keyword) arguments into a dictionary. Additionally, existing dictionaries can be unpacked directly into a function, allowing their key-value pairs to be passed as named arguments in a flexible way.

## Task 4:
Mixing *args in Function Calls:


```python
def f(*args):
    for i in args:
        print(i)
f(*[1, 2, 3], *[4, 5, 6])
```

We observe that you can pass multiple unpacked lists into *args, and all the values from these lists are combined into a single tuple. This allows the function to handle multiple sequences of values in a unified and flexible way.


## Task 5:
Mixing **kwargs in Function Calls:


```python
def f(**kwargs):
    for k, v in kwargs.items():
        print(k, '->', v)
f(**{'a': 1, 'b': 2}, **{'x': 3, 'y': 4})
```

We observe that multiple dictionaries can be unpacked into a single **kwargs. When this is done, all key-value pairs from the dictionaries are collected into one dictionary, allowing the function to access them as named arguments in a flexible manner.

## Task 6:
Using Annotations:


```python
def f(a: int, b: str) -> float:
    print(a, b)
    return 3.5
f(1, 'foo')
print(f.__annotations__)
```

We observe that annotations serve as type hints, not strict restrictions. They indicate the expected input and output types of a function, helping to improve code readability and maintainability without enforcing type checking at runtime.

## Task 7:
Annotations with Default Values:


```python
def f(a: int = 12, b: str = 'baz') -> float:
    print(a, b)
    return 3.5
print(f.__annotations__)
f()
```

We observe that default values can be used alongside annotations in a function. While annotations provide type hints for expected inputs and outputs, they do not enforce any type rules at runtime, allowing flexibility in the values passed to the function.

## Task 8:
Annotations Ignored by Python:


```python
def f(a: int, b: str) -> float:
    print(a, b)
    return 1, 2, 3
f('foo', 2.5)
```

We observe that function annotations do not enforce types. Even if the values passed to a function do not match the annotated types, Python will not raise any errors, allowing you to break the suggested type rules without any runtime complaints.
