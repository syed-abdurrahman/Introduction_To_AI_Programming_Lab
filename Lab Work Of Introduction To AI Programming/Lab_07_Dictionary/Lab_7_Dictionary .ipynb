## **AIM**
The aim of this lab is to practice Python dictionaries for storing, accessing, updating, and manipulating structured data. Students will learn to handle nested dictionaries, check for keys, use dictionary methods such as `.keys()`, `.values()`, `.items()`, `.pop()`, `.get()`, and perform real-world operations like monitoring solar home systems, student records, weather data, inventory management, and patient monitoring. This lab demonstrates the usefulness of dictionaries for flexible, key-based data management in AI and engineering applications.

------


```python
#  Smart Home Energy Meter
smart_meter = {
    "time": "18:32",
    "grid_power": 720,
    "solar_power": 340,
    "battery_soc": 67,
    "appliance_status": {"AC": "OFF", "Fridge": "ON", "Motor": "OFF"}
}
```

Q1: Print the "grid_power" value

```python
print("Grid Power:", smart_meter["grid_power"])
```

**Q2:** Add a new appliance "TV": "ON" inside "appliance_status"


```python
smart_meter["appliance_status"]["TV"] = "ON"
print(smart_meter["appliance_status"])
```

**Q3:** Change "battery_soc" to 72


```python
smart_meter["battery_soc"] = 72
print(smart_meter["battery_soc"])
```

**Q4:** Check if "solar_power" exists in the dictionary


```python
print("solar_power exists:", "solar_power" in smart_meter)
```

**Q5:** Delete the "Motor" key from "appliance_status"


```python
smart_meter["appliance_status"].pop("Motor")
print(smart_meter["appliance_status"])
```

# Task 3: Weather Monitoring Station



```python

weather = {
    "hour": 14,
    "temperature": 33.8,
    "humidity": 52,
    "wind": {"speed": 12, "direction": "NE"},
    "alerts": []
}
```

**Q2:** Print The Wind Direction


```python

print("Wind Direction:", weather["wind"]["direction"])
```

**Q2:** Add a new alert "High Temperature"


```python
weather["alerts"].append("High Temperature")
print(weather["alerts"])
```

**Q3:** Change the wind speed to 15


```python
weather["wind"]["speed"] = 15
print(weather["wind"])
```

**Q4:** Use `.keys()` to list all top-level keys


```python
print("Top-level keys:", weather.keys())
```

**Q5:** Remove "humidity" using `.pop()`


```python
weather.pop("humidity")
print(weather)
```

# Task 4: Inventory System for Electronics Shop


```python
inventory_item = {
    "item_id": 1042,
    "name": "PWM Solar Charge Controller",
    "stock": 18,
    "price": 2500,
    "features": {"voltage": "12/24V", "current": "20A", "brand": "SolarTech"}
}
```

# Task 5: Hospital Patient Monitoring System



```python
patient = {
    "id": 907,
    "name": "Faisal",
    "heart_rate": 82,
    "oxygen_level": 97,
    "medications": {"aspirin": "8AM", "vitamin_D": "12PM"}
}
```

**Q1:** Print the oxygen level


```python

print("Oxygen Level:", patient["oxygen_level"])
```

**Q2:** Add a new medication "antibiotic": "6PM"


```python
patient["medications"]["antibiotic"] = "6PM"
print(patient["medications"])
```

**Q3:** Update heart rate to 78


```python
patient["heart_rate"] = 78
print(patient["heart_rate"])
```

**Q4:** Get the medication schedule for "insulin" using `.get()` with default "Not Prescribed"


```python
print("Insulin schedule:", patient["medications"].get("insulin", "Not Prescribed"))
```

**Q5:** Use `.items()` to print all key-value pairs


```python
print("Patient Data Items:", patient.items())
```
