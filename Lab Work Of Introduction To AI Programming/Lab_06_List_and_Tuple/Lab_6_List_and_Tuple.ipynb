# Task 6: List Comprehensions & Conversion

**Q1:** Convert a tuple to a list


```python
values = (2, 4, 6, 8, 10)
values_list = list(values)
print(values_list)
```

**Q2:** Use a list comprehension to square each value


```python
squared = [x**2 for x in values_list]
print(squared)
```

**Q3:** Filter squared values > 30


```python
filtered = [x for x in squared if x > 30]
print(filtered)
```
