# **AIM**

The aim of this lab is to understand and implement for loops and while loops in Python to perform repetitive tasks such as data printing, calculations, simulations, and condition-based execution in engineering-related problems.


----

## 1. For Loop – Task 1

A lab keeps records of resistor values. Students need to print each resistance value for verification.
Print each resistance value from a list of resistors.
Create a list, then use for loop to display each item.


```python
resistors = [10, 22, 47, 100]
for r in resistors:
    print(r)
```

------

## 2. For Loop – Task 2

Current readings are stored in a list. Engineers want to calculate the square of each value for power loss calculation.
Print the square of each current reading using formula I².
Make a list of current values, loop and multiply each by itself.


```python
currents = [1, 2, 3, 4]
for i in currents:
    print(i ** 2)
```

------

## 3. For Loop – Task 3

Instruments in a lab are labelled by serial numbers. Students want to auto-generate these numbers.
Print serial numbers from 1 to 10 for lab instruments.
Use range(1, 11) in for loop.


```python
for serial in range(1, 11):
    print(serial)
```

## 4. For Loop – Task 4

A power company records voltage values at different times of the day.
Print each voltage value from a list of voltages.


```python
voltages = [220, 215, 230, 225]
for v in voltages:
    print(v)
```

## 5. For Loop – Task 5

A lab has capacitor values stored in microfarads.
Print double of each capacitor value from a list.


```python
capacitors = [10, 22, 47]
for c in capacitors:
    print(c * 2)
```

## 6. For Loop – Task 6

A lab records temperature readings from 5 different sensors.
Print all sensor temperatures from a list.


```python
temperatures = [25, 27, 26, 28, 24]
for t in temperatures:
    print(t)
```

## 7. For Loop – Task 7

A solar system records daily energy in kWh.
Print half of each solar energy reading from a list.


```python
energy = [10, 12, 8, 15]
for e in energy:
    print(e / 2)
```

#  While Loop
## 8. While Loop – Task 1

During charging, the battery slowly increases its percentage.
Keep printing "Charging..." until battery percentage reaches 100%.
Start with battery = 0, increase by 10 in each loop.


```python
battery = 0
while battery < 100:
    print("Charging...", battery)
    battery += 10
```

------

## 9. While Loop – Task 2

Engineers use countdown timers for testing.
Countdown timer from 10 to 0, showing "Test Running".


```python
timer = 10
while timer >= 0:
    print(timer, "Test Running")
    timer -= 1
```

---------

## 10. While Loop – Task 3

A voltage monitoring system asks for user input repeatedly.
Keep asking user for voltage until they enter 0.


```python
voltage = -1
while voltage != 0:
    voltage = int(input("Enter voltage: "))
```

-------

## 11. While Loop – Task 4

Power in a system doubles when load increases.
Keep doubling the power until it exceeds 1000 W.
Start power = 1.


```python

power = 1
while power <= 1000:
    print(power)
    power *= 2
```

------

## 12. While Loop – Task 5

A machine keeps running until its cycle count reaches a limit.
Start from cycle = 1 and keep printing "Cycle Running" until cycle reaches 5.


```python
cycle = 1
while cycle <= 5:
    print("Cycle Running")
    cycle += 1
```

---------

## 13. While Loop – Task 6

A motor runs until its speed reaches 3000 RPM.
Start from 0 and add 500 each loop, printing the speed.


```python
speed = 0
while speed <= 3000:
    print(speed)
    speed += 500
```
