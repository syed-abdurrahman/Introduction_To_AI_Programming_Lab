## 1. Voltage Level Classification (Read + Write File)

A text file contains circuit voltage readings (one per line).

 Task: Read the file and classify each voltage as: 
 1. Low (< 110V) 
 2. Medium (110V–220V) 
 3. High (> 220V)
 
 Save the classification results into a new file.

```python
with open("Lab_3_voltages.txt", "r") as file:
    readings = file.readlines()
    print(readings)
```

```python
results = []  # store classified results

for line in readings:

    voltage = float(line.strip())  # convert string to float
    if voltage < 110:Level = "Low"
    elif 110 <= voltage <= 220:Level = "Medium"
    else:Level= "High"
    results.append((voltage, Level))
results
```

```python
with open("Lab_3_classified_voltages.txt", "w") as f:
    for voltage,Level in results:
        f.write(f"{voltage}V , {Level}\n")
```

## 3. Task: Current Rating Check (Electrical Machines)
A file has current values (in Amperes).

Task: Check for each line: 
1. If current < 5A → Safe 
2. If 5A ≤ current < 10A → Warning 
3. If ≥ 10A → Overload

```python
with open("Lab_3_current.txt") as file:
    for line in file:
        current = float(line.strip())

        if current < 5:
            print(current, "A → Safe")
        elif current < 10:
            print(current, "A → Warning")
        else:
            print(current, "A → Overload")
```

## 4. Task: File Extension Validator (General Engineering Reports)
A directory has multiple engineering reports.

Task: Write a program to scan all files. 
1. If file ends with .pdf → Valid Report 
2. If file ends with .docx → Needs Conversion 
3. Otherwise → Invalid Format

Count and display the numbers of each type.

```python
import os

def validate_reports(directory):
    # Counters
    valid_count = 0
    convert_count = 0
    invalid_count = 0

    # Scan all files in the give path means that below i provide my file with thw name of Introduction To AI
    for file in os.listdir(directory):
        if file.endswith(".pdf"):
            print(f"{file} : Valid Report")
            valid_count += 1
        elif file.endswith(".docx"):
            print(f"{file} : Needs Conversion")
            convert_count += 1
        else:
            print(f"{file} : Invalid Format")
            invalid_count += 1

    # Summary
    print("\nSummary:")
    print(f"Valid Reports (.pdf): {valid_count}")
    print(f"Needs Conversion (.docx): {convert_count}")
    print(f"Invalid Format (others): {invalid_count}")



directory_path = "C:/EngineeringReports"  
validate_reports(r"C:\Users\Syed Abdur Rahman\Desktop\3rd Semester\Introduction To AI")
```

## 13. Task : Smart Lighting Based on Lux Value
Smart buildings adjust lighting automatically based on sensor lux values. We need to classify light levels. Check lux value and print lighting condition:

• If lux < 50 → "Pitch Dark – Lights ON full"

• Elif lux < 100 → "Very Dim – Lights ON medium"

• Elif lux < 300 → "Dim Light – Lights ON low"

• Elif lux < 600 → "Normal Brightness – Lights Balanced"

• Elif lux < 1000 → "Bright – Lights OFF"

• Else → "Too Bright – Close Curtains"

• Store lux value, use multiple elif steps.

```python
lux = 250

if lux < 50:
    print("Pitch Dark – Lights ON full")
elif lux < 100:
    print("Very Dim – Lights ON medium")
elif lux < 300:
    print("Dim Light – Lights ON low")
elif lux < 600:
    print("Normal Brightness – Lights Balanced")
elif lux < 1000:
    print("Bright – Lights OFF")
else:
    print("Too Bright – Close Curtains")
```
