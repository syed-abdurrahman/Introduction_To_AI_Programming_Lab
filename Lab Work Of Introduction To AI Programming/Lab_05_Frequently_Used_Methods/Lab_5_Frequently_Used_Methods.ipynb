## Task 1: enumerate()
**Q1:** A list contains voltage readings from 5 measurement points. Display the reading number with each voltage.  
**Q2:** A student records resistance values of resistors in a lab. Print the resistor number and its value.


```python
# Voltage readings
voltages = [220, 230, 210, 240, 225]
for i, v in enumerate(voltages, 1):
    print(f"Reading {i}: {v} V")

# Resistance values
resistors = [100, 220, 330]
for A, r in enumerate(resistors, 1):
    print(f"Resistor {A}: {r} Ω")
```

## Task 2: zip()
**Q1:** Combine current and voltage readings and display as pairs.  
**Q2:** Combine component names with their ratings.


```python
currents = [2, 5, 10]
voltages = [220, 230, 240]
for i, v in zip(currents, voltages):
    print("Current:", i, "A, Voltage:", v, "V")

components = ['Resistor', 'Capacitor', 'Inductor']
ratings = ['220Ω', '10µF', '5mH']
for comp, rate in zip(components, ratings):
    print(comp, "->", rate)
```

## Task 3: map()
**Q1:** Convert a list of temperatures from Celsius to Fahrenheit.  
**Q2:** Convert a list of voltages in volts to kilovolts.


```python
# Celsius to Fahrenheit
temps_c = [0, 25, 100]
temps_f = list(map(lambda x: (x*9/5)+32, temps_c))
print("Fahrenheit:", temps_f)

# Volts to kilovolts
voltages = [220, 1100, 3300]
kv = list(map(lambda x: x/1000, voltages))
print("kV:", kv)
```

## Task 4: filter()
**Q1:** Keep only current readings greater than 10 A.  
**Q2:** Select capacitors above 50 µF.


```python
# 1. A list contains current readings. Use filter() to keep only readings greater than 10 A.
I_Reading=[2,3,4,5,12,13,14,15]
current_reading = list(filter(lambda x: x > 10, I_Reading))
print(current_reading)

# 2.	A capacitor bank has values [10, 20, 50, 100]. Use filter() to select capacitors above 50 µF.
capacitor=[10, 20, 50, 100]
C_above_50=list(filter(lambda x:x>=50,capacitor))
print(C_above_50)
```

## Task 10: all()
**Q1:** Check if all resistors have resistance above 100 ohms.  
**Q2:** Check if all solar panels are active (True).


```python
resistors = [150, 200, 120]
print("All resistors >100Ω:", all(r > 100 for r in resistors))

solar_panels = [True, True, True]
print("All solar panels active:", all(solar_panels))
```

## Task 16: abs()
**Q1:** Find the absolute value of alternating current readings.  
**Q2:** Get the magnitude of a negative voltage reading.


```python
ac_readings = [-5, 10, -15]
abs_readings = list(map(abs, ac_readings))
print("Absolute AC readings:", abs_readings)

voltage = -220
print("Magnitude of voltage:", abs(voltage))
```

## Task 19: pow()
**Q1:** Print the square of current (I²) for a list of current values.


```python
currents = [2, 3, 4]
squared_currents = list(map(lambda x: pow(x, 2), currents))
print("Current squared (I²):", squared_currents)
```
