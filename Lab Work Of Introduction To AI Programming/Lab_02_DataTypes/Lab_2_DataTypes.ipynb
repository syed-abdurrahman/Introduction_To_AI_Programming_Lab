## 1. String Formatting

```python
resistance = 10
current = 2.5324
voltage = resistance * current
```

## 2. File Handling: Read CSV & Add 2 to Each Row

## 3. Count Integer Values in CSV (NO if-statements)

```python
import csv

with open("Lab_2_integers.csv") as file:
    nums = [float(row[0]) for row in csv.reader(file)]

count_integers = sum(x.is_integer() for x in nums)

print("Total integers:", count_integers)
```

## 4. Bit Parity Checker

```python
signal = 0b11010110

ones = signal.bit_count()

parity = "Even" if ones % 2 == 0 else "Odd"

print("Binary Signal:", bin(signal))
print("Number of 1s:", ones)
print("Parity:", parity)
```

## 5. Represent Impedance Using Complex Numbers

```python
import cmath
import math

R = 50
L = 0.1
C = 100e-6
f = 60
w = 2 * math.pi * f

Z_R = complex(R, 0)
Z_L = complex(0, w * L)
Z_C = complex(0, -1 / (w * C))

print("Resistor Impedance:", Z_R)
print("Inductor Impedance:", Z_L)
print("Capacitor Impedance:", Z_C)
```

-----------

## 6. Find Total Impedance

```python
Z_total = Z_R + Z_L + Z_C

print("Total Impedance:", Z_total)
print("Real Part:", Z_total.real)
print("Imag Part:", Z_total.imag)
print("Magnitude:", abs(Z_total))
print("Phase:", cmath.phase(Z_total))
```

----------

## 7. Compute Circuit Current

```python
V = complex(120, 0)
I = V / Z_total

print("Current Magnitude:", abs(I))
print("Current Phase:", cmath.phase(I))
```

## 8. Run and Understand

### (a) Ohm’s Law

```python
V = 230
I = 5
R = V / I

print(f"The voltage is {V} V, the current is {I} A, so the resistance is {R:.2f} Ω.")
print("The voltage is {} V, the current is {} A, so the resistance is {:.2f} Ω."
      .format(V, I, R))
```

### (b) AC Power

```python
P = 2000
V = 220
I = P / V

print(f"The real power is {P} W at {V} V supply, drawing {I:.2f} A current.")
print("The real power is {0} W at {1} V supply, drawing {2:.2f} A current."
      .format(P, V, I))
```

### (c) RLC Circuit Impedance

```python
R = 50
X_L = 30
X_C = 20

Z = complex(R, X_L - X_C)

print(f"Impedance of the circuit is {Z.real} + j{Z.imag} Ω.")
print("Impedance of the circuit is {0} + j{1} Ω."
      .format(Z.real, Z.imag))
```

### (d) Efficiency Calculation

```python
P_out = 950
P_in = 1000
eff = (P_out / P_in) * 100

print(f"The machine efficiency is {eff:.1f}% when output is {P_out} W and input is {P_in} W.")
print("The machine efficiency is {:.1f}% when output is {} W and input is {} W."
      .format(eff, P_out, P_in))    
```
