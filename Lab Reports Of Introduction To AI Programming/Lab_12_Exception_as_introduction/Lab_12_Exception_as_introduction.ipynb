## Task 1: Handle Invalid User Input

### Problem  
The program asks the user to enter numeric values for total marks and obtained marks.  
Users may enter invalid input such as letters or symbols.  
The program must not crash and should display a clear error message.

### Concepts Used  
- `try`  
- `except`  
- `ValueError`

```python
try:
    total_marks = int(input("Enter total marks: "))
    obtained_marks = int(input("Enter obtained marks: "))
except ValueError:
    print("Error: Please enter numeric values only.")
```

------

## Task 2: Handle Division by Zero

### Problem  
If total marks are entered as zero, calculating the percentage will cause an error.  
Modify the program to handle this case safely.

### Concepts Used  
- `ZeroDivisionError`  
- Multiple `except` blocks

```python
try:
    total_marks = int(input("Enter total marks: "))
    obtained_marks = int(input("Enter obtained marks: "))
    percentage = (obtained_marks / total_marks) * 100
except ValueError:
    print("Error: Invalid input. Please enter numbers only.")
except ZeroDivisionError:
    print("Error: Total marks cannot be zero.")
```

---------

## Task 3: Validate Logical Conditions Using raise

### Problem  
Obtained marks must not be negative and must not exceed total marks.  
If this rule is violated, raise an exception manually.

### Concepts Used  
- `raise`  
- `Exception`

```python
try:
    total_marks = int(input("Enter total marks: "))
    obtained_marks = int(input("Enter obtained marks: "))

    if obtained_marks < 0 or obtained_marks > total_marks:
        raise Exception("Error: Obtained marks must be between 0 and total marks.")

    percentage = (obtained_marks / total_marks) * 100
except Exception as e:
    print(e)
```

--------

## Task 4: Execute Code Only When No Error Occurs

### Problem  
Display the calculated percentage and pass/fail status only if no exception occurs.

### Concepts Used  
- `try`  
- `except`  
- `else`

```python
try:
    total_marks = int(input("Enter total marks: "))
    obtained_marks = int(input("Enter obtained marks: "))

    if total_marks == 0:
        raise ZeroDivisionError("Total marks cannot be zero.")

    if obtained_marks < 0 or obtained_marks > total_marks:
        raise Exception("Invalid marks entered.")

    percentage = (obtained_marks / total_marks) * 100

except ValueError:
    print("Error: Enter numeric values only.")
except ZeroDivisionError as e:
    print(e)
except Exception as e:
    print(e)
else:
    print("Percentage:", percentage)
    if percentage >= 50:
        print("Status: Pass")
    else:
        print("Status: Fail")
```

--------

## Task 5: Guarantee Program Completion Using finally

### Problem  
No matter what happens during execution, the program must display:  
**Result processing completed.**

### Concepts Used  
- `finally`

```python
try:
    total_marks = int(input("Enter total marks: "))
    obtained_marks = int(input("Enter obtained marks: "))

    if total_marks == 0:
        raise ZeroDivisionError

    if obtained_marks < 0 or obtained_marks > total_marks:
        raise Exception("Invalid marks entered.")

    percentage = (obtained_marks / total_marks) * 100

except ValueError:
    print("Error: Please enter valid numbers.")
except ZeroDivisionError:
    print("Error: Total marks cannot be zero.")
except Exception as e:
    print(e)
else:
    print("Percentage:", percentage)
    print("Result:", "Pass" if percentage >= 50 else "Fail")
finally:
    print("Result processing completed.")
```
