## **Aim of the Lab**

The aim of this laboratory experiment is to calculate the **total impedance** of an electrical circuit by adding multiple complex impedances. The experiment focuses on determining the **real (resistive)** and **imaginary (reactive)** components of the total impedance. Additionally, the **magnitude** and **phase angle** of the total impedance are calculated using the `abs()` and `cmath.phase()` functions to analyze the AC circuit behavior.

## Task: Total Impedance

i. Add the three impedances. 

ii. Print real part (resistive) and imaginary part (reactive). 

iii. Print magnitude and phase angle (use abs() and cmath.phase()).

```python
import cmath
import math

# Given circuit parameters
R = 50
L = 0.1
C = 100e-6
f = 60
w = 2 * math.pi * f
```

```python
# Impedances
Z_R = complex(R, 0)
Z_L = complex(0, w * L)
Z_C = complex(0, -1 / (w * C))
```

```python
# Total impedance
Z_total = Z_R + Z_L + Z_C

print("Total Impedance:", Z_total)
```

```python
# Real and imaginary parts
print("Resistive Part (Real):", Z_total.real)
print("Reactive Part (Imag):", Z_total.imag)
```

```python
# Magnitude & Phase
print("Magnitude:", abs(Z_total))
print("Phase (radians):", cmath.phase(Z_total))
```
