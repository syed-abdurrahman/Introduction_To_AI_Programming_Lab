## **Aim of the Lab**

The aim of this experiment is to understand and apply the built-in `sum()` function in Python to calculate total values from a list. The experiment focuses on finding the **total power consumed by electrical components** and determining the **equivalent resistance of resistors connected in series** by adding their respective values.

### **Tasks**

The built-in `sum()` function is used to add all numerical values in a list.

1. The power consumed by individual components is stored in a list.  
   Calculate the **total power consumed** by adding all the readings using the `sum()` function.

2. A list contains the resistance values of resistors connected in **series**.  
   Find the **equivalent resistance** by summing all resistance values.



```python
#  Total power consumed
power = [10, 15, 20, 5]
total_power = sum(power)
print("Total Power:", total_power)
```

```python
#  Equivalent resistance
resistances = [5, 10, 15]
equiv_resistance = sum(resistances)
print("Equivalent Resistance:", equiv_resistance)
```
