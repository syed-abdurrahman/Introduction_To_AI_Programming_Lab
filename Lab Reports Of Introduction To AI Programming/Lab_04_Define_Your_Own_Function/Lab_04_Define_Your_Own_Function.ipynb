## **Aim of the Lab**

The aim of this experiment is to design a function that calculates the **total resistance of resistors connected in series**. The function accepts a list or tuple of resistance values, returns their sum, and uses **docstrings** and **type annotations** to improve code readability and documentation. Additionally, the experiment demonstrates how to access and display function annotations using the `__annotations__` attribute.

# **Task**
In circuit design, calculating **series resistance** is an important requirement.

- Write a function `series()` that takes a **list or tuple of resistances** as input.
- The function should **return the total resistance** by adding all values.
- Add appropriate **docstrings** to explain the function’s purpose and parameters.
- Use **type annotations** for the function parameters and return value.
- Print the function’s `__annotations__` to display the annotated types.




```python
def series(resistances):
    """
    Calculate total series resistance............
    """
    return sum(resistances)

# Print annotations
print(series.__annotations__)

print("series([10, 20, 30]) ->", series([10, 20, 30]), "ohms")
print("series((1.5, 2.5, 3.0)) ->", series((1.5, 2.5, 3.0)), "ohms")
print("series([]) ->", series([]), "ohms")
 
```
