## **Aim of the Lab**

The aim of this experiment is to design a simple student management system using Python data structures. The experiment focuses on storing student names, roll numbers, and grades using lists, tuples, and nested lists. It also aims to demonstrate the difference between mutable and immutable data types by allowing grade updates while keeping roll numbers fixed.

## **Scenario**

You are designing a small **Student Management System** to maintain student data for a class.  
Each student has:
- A **name**
- A **roll number**
- A **list of grades**

The system allows **updating grades** while keeping **roll numbers fixed**.

---

**Q1:** Create a list of student names: `["Ali", "Usman", "Hassan", "Kamran"]`


```python
names = ["Ali", "Usman", "Hassan", "Kamran"]
print(names)
```

**Q2:** Create a tuple of student roll numbers: `(101, 102, 103, 104)`


```python
roll_numbers = (101, 102, 103, 104)
print(roll_numbers)
```

**Q3:** For each student, create a nested list of grades (3 subjects)


```python
grades = [[85, 90, 78], [88, 92, 80], [80, 85, 90], [75, 80, 82]]
print(grades)
```

**Q4:** Update Ali’s second grade to 95


```python
grades[0][1] = 95
print("Updated Ali's grades:", grades[0])
```

**Q5:** Add a fourth grade for Hassan


```python
grades[2].append(88)
print("Updated Hassan's grades:", grades[2])
```

**Q6:** Attempt to change Kamran’s roll number and explain the outcome


```python
# roll_numbers[3] = 105  # This will raise TypeError because tuples are immutable
```

**Q7:** Print all student names with their updated grades in a readable format


```python
for i, name in enumerate(names):
    print(f"{name} (Roll No: {roll_numbers[i]}), Grades: {grades[i]}")
```

In this experiment, a basic student management system was successfully implemented using Python data structures.
