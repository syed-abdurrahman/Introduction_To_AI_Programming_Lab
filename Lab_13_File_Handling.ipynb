## **Aim of the Lab**
The aim of this laboratory experiment is to practice reading from, writing to, and appending data in **CSV files** using Python. The experiment focuses on creating a simple energy consumption logging system for a residential building, storing daily energy usage in a CSV file, and performing operations such as writing new data, reading existing records, and appending additional entries while maintaining the CSV format.

## **Lab Tasks:** 

## Task 1: Create and Write Energy Data to a CSV File

You are developing a **simple energy consumption logging system** for a residential building.  
Daily electricity usage is stored in a **CSV file** so it can be easily read, updated, and organized.

**Question:**  
Create a CSV file named `Lab_13_energy_data.csv`.  
Write a header and some daily energy consumption records.  
Each record must be written on a new line following the header.

---

```python
import csv

# Data to write
records = [
    ["Date", "Energy Consumed (kWh)"],
    ["2025-09-01", 120],
    ["2025-09-02", 135],
    ["2025-09-03", 128]
]
```

```python
# Create and write CSV file
with open("Lab_13_energy_data.csv", mode="w", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(records)

print("Lab_13_energy_data.csv created and data written successfully.")
```

-------

## Task 2: Read and Display CSV File Data

**Question:**  
Read the contents of `energy_data.csv` and display all records in a readable format.  
The data must be read from the file, not hardcoded.

**Answer:**


```python
import csv

with open("Lab_13_energy_data.csv", mode="r") as file:
    reader = csv.reader(file)
    for row in reader:
        print(row)
```

---------

## Task 3: Append New Energy Records

**Question:**  
Append new daily energy consumption records to `energy_data.csv` without removing existing data.

**Answer:**

```python
import csv

new_records = [
    ["2025-09-04", 140],
    ["2025-09-05", 150]
]
```

```python
with open("Lab_13_energy_data.csv", mode="a", newline="") as file:
    writer = csv.writer(file)
    writer.writerows(new_records)

print("New energy records appended successfully.")
```
