## 1Ô∏è‚É£ Install IDE ‚Äì Python / Anaconda

For this lab, it is essential to have a **Python development environment** installed. The recommended options are:

- **Anaconda** ‚Äì A full-featured Python distribution with pre-installed libraries and IDEs. Recommended for beginners and AI programming.  
- **Python (standalone)** ‚Äì Official Python distribution. Requires manual installation of libraries and IDEs.  

---

### üîπ Option 1: Installing Anaconda (Recommended)

**Why choose Anaconda?**

- Comes with Python, Conda package manager, and over 150 pre-installed scientific libraries (NumPy, Pandas, Matplotlib, Scikit-learn, etc.)  
- Includes **Jupyter Notebook** and **Spyder IDE** for interactive programming  
- Simplifies **virtual environment management** to avoid library conflicts  
- Beginner-friendly and widely used in AI/ML courses and research  

**Steps to Install Anaconda:**

1. **Download Anaconda**  

   - Visit the official website: [Anaconda Distribution](https://www.anaconda.com/download)  
   - Select your operating system: **Windows / macOS / Linux**  
   - Download the latest **Python 3.x version**  

2. **Run the Installer**  

   - Windows/macOS: Double-click the installer and follow the instructions  
   - Linux: Open terminal and run:
     ```bash
     bash Anaconda3-<version>-Linux-x86_64.sh
     ```
   - Accept license agreement, choose installation path  

3. **Optional:** Add Anaconda to PATH (recommended for command-line use)  

4. **Verify Installation:**  
   Open **Anaconda Prompt** (Windows) or terminal (macOS/Linux) and run:
   ```bash
   conda --version
   python --version

## 4Ô∏è‚É£ Jupyter Notebook ‚Äì Interactive Python IDE

**Jupyter Notebook** is an open-source, web-based interactive development environment widely used in **Python programming, AI, and data science**. It allows you to write and execute Python code in a **cell-based, interactive format**, making it perfect for experiments, visualizations, and learning.

---

### üîπ Why Use Jupyter Notebook?

- **Interactive Coding:** Write code in cells and execute them independently  
- **Documentation:** Combine code, text, Markdown, images, and equations in a single notebook  
- **Visualization:** Easily display charts, graphs, and plots with libraries like Matplotlib or Seaborn  
- **Experimentation:** Test snippets of code without running the entire script  
- **Learning:** Ideal for students to document labs and assignments  

---

### üîπ Key Features

- **Notebook Cells:** Separate code, markdown, and outputs for better organization  
- **Support for Libraries:** Numpy, Pandas, Matplotlib, Seaborn, Scikit-learn, TensorFlow, PyTorch, etc.  
- **Export Options:** Save notebooks as `.ipynb`, `.html`, or `.pdf`  
- **Interactive Widgets:** Create interactive plots and dashboards  
- **Kernel Management:** Switch Python versions or environments using kernels  

---

## 5Ô∏è‚É£ Virtual Environments ‚Äì Manage Python Projects Efficiently

A **virtual environment** is an isolated workspace for Python projects. It allows you to manage **dependencies, packages, and Python versions** separately for each project, preventing conflicts between libraries.

---

### üîπ Why Use Virtual Environments?

- **Isolation:** Keep project dependencies separate  
- **Avoid Conflicts:** Different projects can use different library versions  
- **Reproducibility:** Easily replicate the environment for deployment or sharing  
- **Safe Testing:** Test new packages without affecting the global Python installation  

---

### üîπ Types of Virtual Environments

1. **Conda Environments (Anaconda users):**  
   - Managed using the **Conda package manager**  
   - Recommended for AI, data science, and ML projects  

2. **venv (Python Standard Library):**  
   - Built-in Python module for creating virtual environments  
   - Lightweight and sufficient for small projects  

---

### üîπ Creating a Virtual Environment

#### Using Conda (Recommended for AI Labs)

```bash
# Create a new environment named "ai_lab" with Python 3.11
conda create -n ai_lab python=3.11

# Activate the environment
conda activate ai_lab

# Deactivate the environment
conda deactivate

# List all environments
conda env list

## 6Ô∏è‚É£ Online Code Editors ‚Äì Run Python Without Installation

Sometimes, you may not have Python installed locally or want to quickly test code. **Online code editors** allow you to write, run, and share Python code directly in your browser.

---

### üîπ Why Use Online Code Editors?

- **No Installation Required:** Start coding immediately without setting up Python  
- **Cross-Platform:** Works on Windows, macOS, Linux, and even mobile devices  
- **Collaboration:** Easily share code with classmates or instructors  
- **Cloud Resources:** Some platforms provide free GPU/TPU access for AI and ML tasks  
- **Version Control Integration:** Many online editors support GitHub integration  

---

### üîπ Recommended Online Code Editors

1. **Google Colab** (Most Popular)  
   - Website: [https://colab.google/](https://colab.google/)  
   - Features:  
     - Free GPU/TPU access  
     - Jupyter Notebook interface  
     - Supports importing files from Google Drive or GitHub  
     - Pre-installed libraries for AI/ML (TensorFlow, PyTorch, OpenCV)  

2. **Repl.it**  
   - Website: [https://replit.com/](https://replit.com/)  
   - Features:  
     - Supports multiple programming languages  
     - Collaborative coding with real-time editing  
     - Free hosting for small projects  

3. **JupyterLite**  
   - Website: [https://jupyter.org/try](https://jupyter.org/try)  
   - Features:  
     - Lightweight browser-based Jupyter environment  
     - No server required, runs entirely in the browser  

---

### üîπ Getting Started with Google Colab

1. Visit [Google Colab](https://colab.google/)  
2. Sign in with your **Google account**  
3. Create a new notebook (**File ‚Üí New Notebook**)  
4. Write Python code in **code cells** and press `Shift + Enter` to execute  
5. Add **Markdown cells** for notes and documentation  
6. Save your notebook in **Google Drive** or export as `.ipynb`  

---

### üîπ Advantages for Students

- Ideal for **AI and Machine Learning labs** when local GPU is unavailable  
- Easy to **share notebooks with instructors or peers**  
- Allows running **complex Python code** without worrying about system setup  

---

### üîπ Tips

- Regularly **save your work** in Google Drive  
- Use **virtual environments** in local IDEs for offline projects, and Colab for quick experiments  
- Explore **free GPU acceleration** for ML model training  
- Combine with GitHub to **sync your projects online**  

---

**Outcome:**  
Students should be able to **write, run, and share Python code online** using platforms like Google Colab without needing local installations.
